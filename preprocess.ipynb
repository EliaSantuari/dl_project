# ANNOTATIONS UNDERSTANDING
In questo notebook provo a vedere se le annotations date dal professore fanno riferimento al set di train di celeba o di test

```python
import torch
from pathlib import Path
from torchvision.datasets import CelebA
```

```python
data_root = Path("./datasets")
```

```python
celeba_train = CelebA(root=data_root, split="train", download=False)
celeba = CelebA(root=data_root, split="test", download=False)
```

```python
print(len(celeba_train))
print(len(celeba))
```

```python
import matplotlib.pyplot as plt

def show_celeb_image(celeba_r, idx):
    img, attrs = celeba_r[idx]

    attr_names = (celeba_r.attr_names[i] for i, val in enumerate(attrs) if val == 1)

    plt.figure()
    plt.imshow(img)
    plt.title(f"Attributes of the image {idx} of celebA:\n{', '.join(attr_names)}", fontsize=10)
    plt.show()

show_celeb_image(celeba, 2000)
show_celeb_image(celeba_train, 2000)
```

```python
import json
```

```python
annotations_path = Path("./celeba_evaluation.json")

with open(annotations_path, "r") as f:
    annotations = json.load(f)
```

```python
print(annotations[1]["query"])
print(annotations[1]["ground_truth"])
```

```python
show_celeb_image(celeba_train, 7)
print(annotations[1]["query"])
show_celeb_image(celeba_train, 8671)
```

```python
show_celeb_image(celeba, 7)
print(annotations[1]["query"])
show_celeb_image(celeba, 8671)
```

Come posso vedere, per questa determinata query, i risultati utilizzando il set di train non hanno senso, mentre per quanto riguarda il set di test, i risultati sono quelli corretti. Ha senso in quanto i test per la valutazione si devono fare su set di test, non su quelli di train
